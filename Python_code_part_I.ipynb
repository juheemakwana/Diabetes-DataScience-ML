# Modifying and cleanning the values to be more suitable to ML algorithms

Reference:
https://www.cdc.gov/brfss/annual_data/2015/pdf/codebook15_llcp.pdf

For DIABETE3:
((Ever told) you have diabetes (If "Yes" and respondent is female, ask "Was this only when you were pregnant?". If Respondent says pre-diabetes or borderline diabetes, use response code 4.))

- 0 is for no diabetes(2) or only during pregnancy(3)

- 1 is for pre-diabetes or borderline diabetes(4)

- 2 is for yes diabetes(1)

- Remove all 7 (dont know) and Remove all 9 (refused)

```python
# frequency of each unique value
frequency = df_selected['DIABETE3'].value_counts()
print(frequency)
```

```python
df_selected['DIABETE3'] = df_selected['DIABETE3'].replace({2 : 0, 3 : 0, 4 : 1, 1 : 2})
df_selected = df_selected[df_selected.DIABETE3 != 7]
df_selected = df_selected[df_selected.DIABETE3 != 9]
df_selected.DIABETE3.unique()
```

(Adults who have been told they have high blood pressure by a doctor, nurse, or other health professional)

```python
df_selected['_RFHYPE5'] = df_selected['_RFHYPE5'].replace({1 : 0, 2 : 1})
df_selected = df_selected[df_selected._RFHYPE5 != 9]
df_selected._RFHYPE5.unique()
```

#Binary Classification of diabtes vs no diabetes

0: No diabetes & pre-diabetes

1: Diabetes

```python
df_binary = df
df_binary['Diabetes_012'] = df_binary['Diabetes_012'].replace({1 : 0})
df_binary['Diabetes_012'] = df_binary['Diabetes_012'].replace({2 : 1})

df_binary = df_binary.rename(columns = {'Diabetes_012':'Diabetes_binary'})
df_binary.Diabetes_binary.unique()
```

```python
df_binary.head()
```

```python
df_binary.shape
```

```python
df_binary.groupby(['Diabetes_binary']).size()
#Class imbalance
```

```python
#Separate the 0(No Diabetes/Pre-diabetes) and 1(Diabetes)
df_5050_0 = df_binary[df_binary['Diabetes_binary'] == 0]
df_5050_1 = df_binary[df_binary['Diabetes_binary'] == 1]
```

```python
df_5050_0.count()
```

```python
df_5050_1.count()
```

```python
#Select the 35346 random cases from the 0 (non-diabetes group) as we already have 35346 cases from the diabetes risk group
df_5050_0_rand1 = df_5050_0.take(np.random.permutation(len(df_5050_0))[:35346])

#Append the 35346 1s to the 35346 randomly selected 0s
df_5050 = pd.concat([df_5050_0_rand1, df_5050_1], ignore_index=True)
```

```python
df_5050.head()
```

```python
df_5050.tail()
```

```python
#Classes are now perfectly balanced
df_5050.groupby(['Diabetes_binary']).size()
```
